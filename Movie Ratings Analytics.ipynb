```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

sns.set_style('white', {'axes.facecolor':'Black'})
#KDE 1 and 2
f, axes = plt.subplots(2, 2, figsize=(15,15))
k1 = sns.kdeplot(movies.Budget, movies.AudienceRatings, shade=True, shade_lowest=False, cmap='inferno', ax=axes[0,0])
k1b = sns.kdeplot(movies.Budget, movies.AudienceRatings, cmap='inferno', ax=axes[0,0])
k2 = sns.kdeplot(movies.Budget, movies.CriticRatings, shade=True, shade_lowest=False, cmap='inferno', ax=axes[0,1])
k2b = sns.kdeplot(movies.Budget, movies.CriticRatings, cmap='inferno', ax=axes[0,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))

#Histogram - Stacked Bars 
list1 = []
labels1 = []
for g in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == g].Budget)
    labels1.append(g)
axes[1,0].hist(list1, stacked = True, bins = 30, rwidth=1, label=mylabels, edgecolor='Black')
legend = axes[1,0].legend()
plt.setp(legend.get_texts(), color='w') #to change the color of the legend.

#Scatterplot
sns.scatterplot(data = movies, x = 'CriticRatings', y = 'AudienceRatings', hue = 'Genre', ax=axes[1,1])
axes[1,1].plot((0,100),(0,100), color='Grey', ls='--')
legend = axes[1,1].legend()
plt.setp(legend.get_texts(), color='w') #to change the color of the legend.

plt.show()
```
