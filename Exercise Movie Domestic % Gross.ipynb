```python
base = pd.read_csv('P4-Section6-Homework-Dataset.csv')
```

```python
base.Genre.unique() #Action, Comedy, Adventure, Animation, Drama
```

```python
base.Studio.unique() #Buena Vista Studios, Sony, Universal, WB, Paramount Pictures, Fox
```

```python
mov = base[(base.Genre == 'action') | (base.Genre == 'comedy') | (base.Genre == 'adventure') | (base.Genre == 'animation') | (base.Genre == 'drama')]
```

```python
movies = mov[(mov.Studio == 'Buena Vista Studios') | (mov.Studio == 'Sony') | (mov.Studio == 'Universal') | (mov.Studio == 'WB') | (mov.Studio == 'Paramount Pictures') | (mov.Studio == 'Fox')]
```

```python
movies.Genre.unique()
```

```python
movies.Studio.unique()
```

```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(figsize=(12,6))
c1 = sns.boxplot(data=movies, y='Gross % US',x = 'Genre', color='gainsboro')
c2 = sns.stripplot(data=movies, y='Gross % US', x='Genre', hue='Studio', alpha=0.6, linewidth=0, size=6)
plt.title('Domestic Gross % by Genre', fontsize=30)
plt.ylabel('Gross % US', fontsize=20)
plt.xlabel('Genre', fontsize=20 )
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0, frameon=False, framealpha=1, )

plt.show()
```
