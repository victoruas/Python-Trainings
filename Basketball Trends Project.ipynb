```python
def myplot3(measure, legend, x, labelX, labelY):
    for name in legend:
        plt.plot(measure[Pdict[name]],ls='--', marker='o',ms=7, label=name)
    plt.xticks(list(range(0,len(x))), x)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xlabel(labelX)
    plt.ylabel(labelY)
    plt.show()
```

```python
print("FTA per Games")
myplot3(FTA / Games ,Players,Seasons,'Seasons', 'FTA per Games')
print("%Accuracy of FT")
myplot3(FT / FTA*100,Players,Seasons,'Seasons', '%Accuracy of FT')
print("Player Playing Style")
myplot3((Points - FT)/FieldGoals,Players,Seasons,'Seasons', 'Player Playing Style')
```
